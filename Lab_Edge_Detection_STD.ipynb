```python
# 2. Custom Sobel (V-H-Both)
hy = np.array([
    [-1,0,1],
    [-2, 0, 2],
    [-1,0,1]
])
hx = np.array([
    [-1,-2,-1],
    [0, 0, 0],
    [1,2,1]
])
img = io.imread('circuit.tif')
# print(img.shape)
img_f1 = convolve2d(img, hy)
# print(img_f1)
img_f2 = convolve2d(img, hx)
# print(img_f2)
img_f3=np.zeros(img.shape)
# print(img_f3.shape)
for i in range(img_f3.shape[0]):
    for j in range(img_f3.shape[1]):
        img_f3[i][j]=math.sqrt(pow(img_f1[i][j],2)+pow(img_f2[i][j],2))


img_f3[(img_f3<80)]=0
show_images([img,img_f1,img_f2,img_f3],['original','y','x','x&y'])
# print(img_f3)
```
